```python
import statsmodels.formula.api as smf
import pandas as pd
import numpy as np

#importing data for various exporting products for the past two decades
data = pd.read_excel("/Users/mr.malhaar/Desktop/ex_data.xlsx", header = 0)
df = pd.DataFrame(data)
df
```

```python
#Setting up the regression to model the relationship between India's GDP and various exporting products

#It's a log-log model in order to make it easy for interpretation as 1% increase in any of the independent variable would
#increase the GDP by the percentage of the coefficient of the independent variable
variables = "np.log(GDP)~np.log(Mineral_Products) + np.log(Chemical_Products) +	np.log(Plastics_and_Rubbers) + np.log(Wood_Products) + np.log(Paper_Goods) + np.log(Textiles) + np.log(Precious_Metals) + np.log(Metals) + np.log(Machines) +	np.log(Transportation) +  np.log(Weapons)"

#performing the regression and then displaying the output
reg = smf.ols(variables, df).fit()
print(reg.summary())
```

```python
#importing the data for India's GNI per capita and employment rate (in services) for the past two decades
data_2 = pd.read_excel("/Users/mr.malhaar/Desktop/emp.xlsx", header = 0)
df_2 = pd.DataFrame(data_2)
df_2
```

```python
#setting up the regression to model the relationship between India's GNI per capita and employment rate
variables_2 = "GNI_per_capita ~ Employment_Rate"
reg = smf.ols(variables_2, df_2).fit()
print(reg.summary())
```

```python
#visualising the regression done above
import seaborn as sns
import matplotlib.pyplot as plt

sns.lmplot(x = 'Employment_Rate' ,y = 'GNI', data = df_2, fit_reg = True, ci = None)

plt.title("Simple Linear Regression")
plt.ylabel("GNI per capita($)")
plt.xlabel("Employment rate in service sector")

plt.show()
```
